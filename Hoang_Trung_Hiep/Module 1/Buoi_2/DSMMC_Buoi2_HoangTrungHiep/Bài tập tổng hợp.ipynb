##**Question 2:**
Viết chương trình phân tích một số nguyên thành các thừa số nguyên tố

Ví dụ: Số 28 được phân tích thành 2 x 2 x 7


```python
import math
def checkSoNguyenTo(n):
  for i in range(2,round(math.sqrt(n))+1):
    if n % i ==0:
      return False
  return True
num = int(input("Nhập vào 1 số nguyên :"))
res =""
while checkSoNguyenTo(num) == False and num != 1:
  for i in range(2,round(math.sqrt(num))+1):
    if num % i ==0 and checkSoNguyenTo(i):
      res+=str(i)+"x"
      num = num//i
      break
if num!=1:
  res+=str(num)
print(res)
```

##**Question 4:**
Viết chương trình liệt kê n số nguyên tố đầu tiên.


```python
import math
def checkSoNguyenTo(n):
  for i in range(2,round(math.sqrt(n))+1):
    if n % i ==0:
      return False
  return True
def lietKeNSoNguyenTo(n):
  i=0
  start = 2
  res=""
  while i!=n:
    if checkSoNguyenTo(start):
      res+=str(start)+""
      start+=1
      i+=1
  return res
print(lietKeNSoNguyenTo(5))
```

##**Question 5:**
Viết chương trình tìm ước số chung lớn nhất, bội số chung nhỏ nhất của hai số tự nhiên a và b.


```python
x = int(input("Nhập vào số nguyên thứ nhất :"))
y = int(input("Nhập vào số nguyên thứ hai :"))
def UCLN(x,y):
  while y:
    x,y = y,x%y
  return x
def BCNN(x,y):
  return x*y//UCLN(x,y)
print(f"UCLN :{UCLN(x,y)}\nBCNN :{BCNN(x,y)}")
```

##**Question 6:**
Dãy số Fibonacci được định nghĩa như sau:

F0 =1, F1 = 1; Fn = Fn-1 + Fn-2 với n>=2.

Hãy viết chương trình tìm số Fibonacci thứ n.


```python
def fibo(n):
  if n==0 or n==1:
    return 1
  return fibo(n-1)+fibo(n-2)
print(fibo(8))
```

##**Question 10:**
Viết chương trình thực hiện chuẩn hoá một xâu ký tự nhập từ bàn phím

(loại bỏ các dấu cách thừa, chuyển ký tự đầu mỗi từ thành chữ hoa, các ký tự khác thành chữ thường)


```python
s = input("Nhập 1 xâu ký tự: ")
def xuLyChuoi(s):
  new_s = ""
  for i in range(len(s)):
      if i == 0 and s[i] == " ":
          continue
      if s[i] == " " and s[i-1] == " ":
          continue
      if "A" <= s[i] <= "Z":
          new_s += chr(ord(s[i]) + 32)
      else:
          new_s += s[i]
  new_s2 = chr(ord(new_s[0]) - 32) + new_s[1:]
  return new_s2
print(xuLyChuoi(s))
```

##**Question 2:**
Viết chương trình tính n! với n là số tự nhiên không âm nhập vào từ bàn phím. Trong đó: 0! = 1 n! = 1.2.3...n


```python
def tinhGiaiThua(n):
  if n == 0 or n==1:
    return 1
  return n*tinhGiaiThua(n-1)
print(tinhGiaiThua(3))
```
