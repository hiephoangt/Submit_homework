```python
#Tổng số tiền
tong_tien = 1000000

#Hàm quay số
import random
def quay_so():
  lst_giai = [random.randint(10,99) for i in range(7)]
  return lst_giai

#Hàm in danh sách giải
def in_danh_sach_giai(danh_sach_giai):
  for i in range(7):
    print(f"Giai {i+1}: {danh_sach_giai[i]}")

#Hàm nhập số lô
def nhap_so_lo():
  while True :
    so_lo = input("Nhập số lô bạn muốn cược (10-99): ")
    if so_lo.isdigit() and 10<=int(so_lo)<=99:
      return int(so_lo)
    else :
      print("Bạn đã nhập sai định dạng. Vui lòng nhập lại!!!!!")

#Hàm nhập tiền cược
def nhap_tien_cuoc():
   while True :
    tien_cuoc = input("Nhập tiền cược (tiền cược < tiền bạn có) : ")
    if tien_cuoc.isdigit() and int(tien_cuoc)<tong_tien:
      return int(tien_cuoc)
    else :
      print("Vui lòng nhập lại tiền cược!!!")

#Chức năng chơi lô
def game_quay_so_lo(so_lo,tien_cuoc):
  global tong_tien

  danh_sach_giai = quay_so()
  in_danh_sach_giai(danh_sach_giai)
  dem_so_giai = 0
  print(f"Số lô bạn chơi là :{so_lo}")
  for i in danh_sach_giai:
    if so_lo == i:
      dem_so_giai+=1
  if dem_so_giai>0:
    tong_tien += dem_so_giai*70
    print('Bạn đã trúng ')
    print(f"Số tiền bạn trúng là :{dem_so_giai*70*tien_cuoc}")
    print(f"Tống số tiền bạn có là :{tong_tien}")
  else :
    tong_tien-= tien_cuoc
    print('Bạn đã thua ')
    print(f"Số tiền bạn thua là :{tien_cuoc}")
    print(f"Tống số tiền bạn có là :{tong_tien}")

#Chức năng nạp tiền
def nap_tien():
  global tong_tien
  while True:
    tien_nap = input("Nhập số tiền bạn muốn nạp: ")
    if tien_nap.isdigit():
      tong_tien += int(tien_nap)
    nap_tien = int(input("Bạn có muốn nạp tiếp 0(không) 1(có): "))
    if nap_tien == 0:
      print(f"Tổng số tiền bạn có : {tong_tien}")
      break
    else :
      continue
#Chức năng rút tiền
def rut_tien():
  global tong_tien
  while True:
    print(f"Số tiền bạn hiện có :{tong_tien}")
    tien_rut = input("Nhập số tiền bạn muốn rút(<số tiền bạn có): ")
    if tien_rut.isdigit() and 0<int(tien_rut)<=tong_tien:
      tong_tien -= int(tien_rut)
    break

#menu
def menu():
  print("GAME LÔ ĐỀ HỌC")
  print("1.Chơi Lô")
  print("2.Nạp tiền")
  print("3.Rút tiền")
  print("4.Thoát game")

#main
from IPython.display import clear_output
def main():
  while True:
    clear_output()
    menu()
    lua_chon = int(input("Nhập lựa chọn: "))
    match lua_chon:
      case 1:
        clear_output()
        while True:
          tien_cuoc = nhap_tien_cuoc()
          so_lo = nhap_so_lo()
          game_quay_so_lo(so_lo,tien_cuoc)
          choi_tiep = int(input("Bạn có muốn chơi tiếp 0(không) 1(có): "))
          clear_output()
          if choi_tiep ==0:
            break

        clear_output(wait =True)
      case 2:
        clear_output()
        nap_tien()
      case 3:
        clear_output()
        rut_tien()
      case 4:
        clear_output()
        break
      case _:
        print("Bạn nhập sai. Vui lòng nhập lại !!!!!!!!!")
        main()

```

```python
main()
```
