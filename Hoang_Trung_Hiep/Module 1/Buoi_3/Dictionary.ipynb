# I. Giới thiệu về Dictionary


**Kiểu dữ liệu Dictionary** là một tập hợp các cặp key-value không có thứ tự, có thể thay đổi và lập chỉ mục (truy cập phần tử theo index).

**Khởi tạo Dictionary** bằng dấu ngoặc nhọn ( {} ) và các phần tử là các cặp key-value sẽ nằm bên trong.

**Mỗi cặp key-value** được xem như là một phần tử, mỗi phần tử cách nhau bởi dấu phẩy ( , )

**Key và value** được phân biệt bởi một dấu hai chấm ( : ), trong đó key phải là duy nhất và value có thể là bất kỳ giá trị nào.

```python
# Khởi tạo một dictionary theo cú pháp phổ biến:
# <name> = {<key_1: value_1>, <key_2: value_2>, ..., <key_n: value_n>}

dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
print(dictCar)
```

# II. Thao tác với Dictionary


## 1.Truy cập các item của Dictionary trong Python

Truy cập các item của Dictionary bằng cách sử dụng ***key*** của nó, bên ***trong dấu ngoặc vuông***, ví dụ:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
print(dictCar["model"])
```

Một cách khác là dùng hàm ***get(\<key\>)*** - hàm này sẽ trả về value của key đã nhập (nếu key không có sẽ trả về None), ví dụ:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
print(dictCar.get("model"))
```

## 2.Thay đổi giá trị của một Dictionary trong Python

Có thể ***thay đổi giá trị*** của một item cụ thể bằng khóa của nó (trong trường hợp khoá ***đã tồn tại***):  

(nếu không, một cặp key-value mới sẽ được tạo)

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar["year"] = 2020
print(dictCar)
```

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    # "year": 1972 (ví dụ key này chưa tồn tại)
}
dictCar["year"] = 2020
print(dictCar)
```

## 3.Duyệt các item của Dictionary trong Python

Có thể duyệt qua một Dictionary bằng cách sử dụng vòng lặp ***for***.

Giá trị trả về sẽ là các key, khi đó có thể dùng hàm get() để lấy value của key tương ứng.

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
for x in dictCar:
    print("Giá trị trả về là key: ", x)
    print("Lấy value của key tương ứng: ", dictCar.get(x))
    print()
```

Khi chỉ muốn lấy các value trong một Dictionary, có thể sử dụng hàm ***values()***, ví dụ:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
for x in dictCar.values():
    print(x)
```

Ngoài ra, còn các hàm tương tự như:

***keys()*** để chỉ lấy các key trong một Dictionary

***items()*** để lấy các cặp item trong một Dictionary

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}

print("Lấy các key: ", dictCar.keys())
print()

print("Lấy các item: ", dictCar.items())
for key, value in dictCar.items():
    print(f"{key} : {value}")
```

## 4.Kiểm tra nếu key tồn tại

Để xác định xem một khóa (key) được chỉ định có ***tồn tại*** trong từ điển hay không, hãy sử dụng từ khóa ***in***:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}

if "model" in dictCar:
    print("Khoa \"model\" co ton tai.")
else:
    print("Khoa \"model\" khong ton tai.")

if "price" in dictCar:
    print("Khoa \"price\" co ton tai.")
else:
    print("Khoa \"price\" khong ton tai.")
```

## 5.Độ dài của một Dictionary trong Python

Để xác định có bao nhiêu item (cặp key-value) trong Dictionary, hãy dùng hàm ***len()***.

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
print(len(dictCar))
```

## 6.Thêm các item vào Dictionary

***Thêm một item*** vào Dictionary được thực hiện bằng cách ***sử dụng khóa mới và gán giá trị*** cho nó:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar["color"] = "yellow"
print(dictCar)
```

## 7.Xóa item của Dictionary

Có nhiều phương pháp để loại bỏ các item của một Dictionary.

### Xóa item của Dictionary bằng hàm pop()
Hàm ***pop()*** xóa item với key được ***chỉ định***:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar.pop("model")
print(dictCar)
```

### Xóa item của Dictionary bằng hàm popitem()
Hàm ***popitem()*** xóa item ***cuối cùng*** (trong các phiên bản trước 3.7, một mục ngẫu nhiên được xóa).

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar.popitem()
print(dictCar)
```

### Xóa item của Dictionary bằng từ khóa del
Lệnh ***del*** sẽ xóa item với key được ***chỉ định***:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
del dictCar["model"]
print(dictCar)
```

### Xóa item của Dictionary bằng hàm clear()
Hàm ***clear()*** xóa ***toàn bộ*** các item của Dictionary.

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dictCar.clear()
print(dictCar)
```

## 8.Copy Dictionary trong Python

**Không thể sao chép từ điển chỉ bằng cách gán dict2 = dict1.**

Vì dict2 sẽ chỉ là một tham chiếu đến dict1 và những thay đổi được thực hiện trên dict1 cũng sẽ xảy ra trên dict2.

Có nhiều cách để tạo một bản sao, một trong các cách đó là sử dụng hàm ***copy()***, ví dụ:

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}
dict1 = dictCar # Sử dụng phép gán =
dict2 = dictCar.copy() # Sử dụng hàm copy()

dictCar["color"] = "yellow" # Thay đổi dictCar

print("dict1: ", dict1) # Bị thay đổi theo
print("dict2: ", dict2)
```

Một cách khác để tạo một bản sao là khởi tạo một Dictionary mới dựa vào dictionary cũ.

```python
dictCar = {
    "brand": "Honda",
    "model": "Honda Civic",
    "year": 1972
}

dict1 = dictCar # Sử dụng phép gán =
dict2 = dict(dictCar) # Khởi tạo một dict mới (dict2) dựa vào dict cũ (dictCar)

dictCar["color"] = "yellow" # Thay đổi dictCar
print("dict1: ", dict1) # Bị thay đổi theo
print("dict2: ", dict2)
```

## 9.Dictionary lồng nhau trong Python

Một Dictionary cũng có thể chứa nhiều Dictionary khác, điều này được gọi là Dictionary lồng nhau, ví dụ:

```python
# Tạo một Dictionary chứa 3 Dictionary khác.

myfamily = {
  "child1" : {  # Dict 1
    "name" : "Đăng",
    "birthday" : 2005
  },
  "child2" : {  # Dict 2
    "name" : "Khôi",
    "birthday" : 2003
  },
  "child3" : {  # Dict 3
    "name" : "Nhã",
    "birthday" : 2003
  },
  "keyz": [1, 2]
}

print(myfamily)
print(myfamily['child1'])
print(myfamily['child1']['name'])
print(myfamily['keyz'][0])
```

Một khai báo khác rõ ràng, dể hiểu hơn. Ví dụ trên tương đương với ví dụ sau:

```python
child1 = {
  "name" : "Van",
  "birthday" : 2004
}
child2 = {
  "name" : "Minh",
  "birthday" : 2007
}
child3 = {
  "name" : "Phuc",
  "birthday" : 2011
}

myfamily = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}

print(myfamily)
```

## 10.Constructor dict() trong Python

Cũng có thể sử dụng constructor dict() để tạo một Dictionary mới:

```python
dictCar = dict(brand="Honda", model="Honda Civic", year=1972)
print(dictCar)

dictCar = dict({
    "brand":"Honda",
    "model":"Honda Civic",
    "year":1972
  })
print(dictCar)
```
